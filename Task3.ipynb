```python
#PAIR PLOT
sns.pairplot(df)
plt.show()
```

```python
#Distribution of numeric variables
df_numeric = df.select_dtypes(include='number')
df_numeric.plot(kind = 'kde',subplots = True, layout = (2,2), sharex = False, figsize = (15,10))
plt.show()
```

```python
#There are many extreme values in Sales and Profit and few in discount and quantity.

#finding out which segment has the highest and lowest profit
p = df.groupby(by ='Segment')[['Profit']].sum()
r = p.sort_values(by = ['Profit'],ascending = True)
r.plot(kind = 'bar',figsize =(9,5),color='magenta')
plt.show()
```

```python
#OBSERVATION 1- From the above observation we can see that Consumer has the highest profit where 
#as Home Office has the lowest profit. So to increase profit we must focus on Home Office and Segment.

## let's compare state vs profit
plt.figure(figsize=(10, 7))
sns.barplot(x="Region", y="Sales", hue="Segment", data=df,palette='rainbow')
plt.show()
```

```python
#OBSERVATION 2- From the above observation we can see that in south region Home Office has 
#highest sale whereas Corporate and consumer have almost equal sale. In West region Consumer 
#has the lowest profit and the other two segments are almost equal. In Central region Corporate 
#has the highest sale whearas Consumer and Home office have almost equal profit. In East region 
#Home office has the highest sale whereas Consumer and Corporate have almost equal sale. In order
#to increase the profit the segments having lowest sale must be taken care of.

plt.figure(figsize =(8,6))
df.groupby(by ='Region')['Profit'].sum().plot(kind = 'pie',autopct='%1.1f%%',explode=(0.1, 0, 0,0),colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue'])
plt.show()
```

```python
#OBSERVATION 3- From the above pie chart it is clearly visible that the Central region has 
#the lowest profit whereas the West has the highest profit. So in order to increase profit 
#we must focus on Central and South region.

plt.figure(figsize=(10, 7))
sns.barplot(x="Region", y="Sales", hue="Category", data=df,palette="rocket_r")
plt.show()
```

```python
#OBSERVATION 4- The categoty wise sales is the highest in South region whereas it is almost similar in other regions.

subcategory_grpby = df.groupby(by ='Sub-Category')[['Sales','Profit']].sum()
sort_profit2 = subcategory_grpby.sort_values(by = ['Profit'],ascending = True)
sort_profit2.plot(kind = 'bar',figsize =(13,7))
plt.show()
```

```python
#OBSERVATION 5- From the above graph we can see that tables has the lowest profit and
#Copiers has the highest profit. The sales is highest in case of phones and lowest in case of Fasteners.

plt.figure(figsize=(15,8))
sns.countplot(x='Sub-Category',data=df,palette='rainbow')
plt.xticks(rotation=90)
plt.show()
```

```python
df.groupby(by ='City')['Profit'].sum().sort_values(ascending = False)[521:].plot(kind = 'bar',color='green')
plt.show()
```

```python
#OBSERVATION 6- In order to increase profit we must focus on Philadelphia since it has the lowest profit.

df.groupby(by ='City')['Profit'].sum().sort_values(ascending = False)[:11].plot(kind = 'bar',color='yellow')
plt.show()
```

```python
#OBSERVATION 7- From the above graph we can see that New York City has the highest profit.

# State vs Profit
plt.figure(figsize =(20,12))
df.groupby(by ='State')['Profit'].sum().sort_values(ascending = True).plot(kind = 'bar')
plt.show()
```

```python
#OBSERVATION 8- From the above graph we can see that sales is lowest in Texas and highest in California.

#Discount vs Profit
plt.figure(figsize=(12,5))
sns.lineplot(x='Discount',y='Profit',data=df)
plt.title("Discount and Profit")
```

```python
#Discount vs Sales
plt.figure(figsize=(12,8))
sns.lineplot(x='Discount',y='Sales',data=df)
plt.title('Discount VS Sales')
```
